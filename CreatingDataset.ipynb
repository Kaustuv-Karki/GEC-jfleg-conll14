# Combining both the datasets

```python
# pos.csv contains the positive sentences which are correct
# Whereas the neg.csv contains all the negative sentences which ate the input so incorrect


negative_sentences = pd.read_csv("neg.csv", dtype="str", names=["input"])
positive_sentences = pd.read_csv("pos.csv", dtype="str", names=["target"])
```

```python
negative_sentences
```

```python
positive_sentences
```

```python
positive_sentences
```

```python
total_dataset_conll = pd.concat([negative_sentences, positive_sentences], axis=1)
total_dataset_conll.columns = ['input', 'target']
```

```python
total_dataset_conll
```

```python


total_dataset = pd.concat([val, valEval, total_dataset_conll], axis = 0)
```

```python
total_dataset.to_csv("total_dataset_jfleg_conll-2014.csv", index = False)
```

```python
eval_total_dataset = total_dataset.sample(800)
```
